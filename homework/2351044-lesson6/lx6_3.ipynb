```python
import pandas as pd
```

```python
st_no = 2351044
```

```python
 # 读取sheet1
df_info = pd.read_excel('xuanke.xls', sheet_name='信息')
df_info
```

```python
# 根据学号筛选数据
st_info = df_info[df_info['学号'] == st_no].iloc[0]
st_info
```

```python
# 设置属性
st_name = st_info['姓名']
st_py = st_info['英文姓名']
st_sex = st_info['性别']
```

```python
# 读取sheet2
df_course = pd.read_excel('xuanke.xls', sheet_name='选课')
st_course = df_course[df_course['学号'] == st_no].iloc[0]
st_course
```

```python
# 获取课程列表
st_course_list = [course for course in st_course.index[1:] if st_course[course] == '√']
st_course_list
```

```python
# 读取sheet3
schedule_df = pd.read_excel('xuanke.xls', sheet_name='上课时间')
schedule_title = schedule_df.columns.tolist()
schedule_list = schedule_df.values.tolist()
schedule_list = [schedule_title] + schedule_list
schedule_list
```

```python
# 替换不在选课列表中的课程
for i in range(1, len(schedule_list)):  # 从1开始，跳过标题行
    for j in range(1, len(schedule_list[i])):  # 从1开始，跳过时间列
        # 检查单元格是否为 NaN 或空值
        if pd.isna(schedule_list[i][j]):
            schedule_list[i][j] = ""  # 如果是空值，则直接替换为空字符串
        else:
            # 拆分课程，处理可能多个课程组合的情况
            courses = schedule_list[i][j].split(',')
            # 替换不在 st_course_list 中的课程
            valid_courses = [course for course in courses if course in st_course_list]
            # 将剩下的课程重新组合，如果没有有效课程则置为 ""
            schedule_list[i][j] = ','.join(valid_courses) if valid_courses else ""

# 打印学生的课表
for row in schedule_list:
    print(row)
st_schedule_list = schedule_list
print(st_schedule_list)
```

```python
print(f"学号：{st_no} 姓名：{st_name} 的课表")
for row in st_schedule_list:
    print('\t'.join([str(cell).ljust(20) for cell in row]))
```

```python
 conflicts = []
# 遍历每个时间段（跳过第一行标题）
for row in st_schedule_list[1:]:
    time_slot = row[0]  # 获取时间段，例如 "上午一二节"
     # 遍历一周的每一天
    for day_idx in range(1, len(row)):
        courses = row[day_idx]  # 获取该时间段某天的课程
        if courses:  # 如果有课程
            course_list = courses.split(',')  # 拆分课程列表
            if len(course_list) > 1:  # 如果课程数量超过 1，说明存在冲突
                conflict_detail = f"{time_slot} {st_schedule_list[0][day_idx]} 有课程冲突: {course_list}"
                conflicts.append(conflict_detail)

conflicts
```
